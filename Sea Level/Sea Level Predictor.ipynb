## 3- Data Analysis 🕵️

#### A- Use matplotlib to create a scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

```python
# Create the scatter plot
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])

# Set labels and title
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')

# x= column year and y = column CSIRO Adjusted Sea Level
```

**Comments:**

- We notice a clear upward trend of the 'CSIRO Adjusted Sea Level' column against the 'Year' column.
- This simply tells us the sea level has been rising as the years progress.

- This upward trend in sea level has important implications for coastal areas and low-lying regions. It indicates that the overall sea level is increasing, which can lead to various impacts such as coastal erosion, increased risk of flooding, and potential damage to coastal ecosystems and infrastructure. The upward trend in sea level is often associated with global climate change and the thermal expansion of seawater, as well as the melting of glaciers and ice sheets.

- Understanding and monitoring the trend in sea level is extremely inportant for assessing and predicting the potential impacts. It also helps in planning appropriate measures to adapt to and mitigate the consequences of rising sea levels.

#### B- Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. 
- Plot the line of best fit over the top of the scatter plot. 
- Make the line go through the year 2050 to predict the sea level rise in 2050.


```python
# Create a scatter plot using Year as x-axis and CSIRO Adjusted Sea Level as y-axis
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], label='Original data')

# Perform linear regression on the entire dataset
slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

# Generate the line of best fit extending to year 2050
line_x = [df['Year'].min(), 2050]
line_y = [intercept + slope * x for x in line_x]

# Plot the line of best fit
plt.plot(line_x, line_y, color='red', label='First line of Best Fit (1880-2013)')

# Set labels and title and legend
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.legend()

# Show the plot
plt.show()

# Predict sea level rise in 2050 based current data
predicted_sea_level_2050 = round((intercept + slope * 2050),2)
print("Predicted Sea Level in 2050:", predicted_sea_level_2050, 'inches')
```

#### C- Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset. 
- Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.
- The x label should be Year, the y label should be Sea Level (**inches**), and the title should be Rise in Sea Level.


```python
# Create a scatter plot using Year as x-axis and CSIRO Adjusted Sea Level as y-axis
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], label='Original data')

# Perform linear regression on the entire dataset
slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

# Generate the line of best fit extending to year 2050
line_x = [df['Year'].min(), 2050]
line_y = [intercept + slope * x for x in line_x]

# Plot the line of best fit
plt.plot(line_x, line_y, color='red', label='First line of Best Fit (1880-2013)')

# Filter the data for the year 2000 onwards
df_recent = df[df['Year'] >= 2000]

# Perform linear regression on the recent data
slope_recent, intercept_recent,r_value_recent, p_value_recent, std_err_recent = linregress(df_recent['Year'], df_recent['CSIRO Adjusted Sea Level'])

# Generate the line of best fit for the recent data extending to year 2050
line_x_recent = [df_recent['Year'].min(), 2050]
line_y_recent = [intercept_recent + slope_recent * x for x in line_x_recent]

# Plot the line of best fit for the recent data
plt.plot(line_x_recent, line_y_recent, color='green', label='Second line of Best Fit (2000-2013)')

# Set labels and title and legend
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
plt.legend()

# Show the plot
plt.show()

# Predict sea level rise in 2050 based on the recent data
predicted_sea_level_2050_recent = round((intercept_recent + slope_recent * 2050),2)
print("Predicted Sea Level in 2050 (2000-2013):", predicted_sea_level_2050_recent, 'inches')
```

**A little bit of learning 📚**

Official documentation of the **linregress** function in the [scipy.stats module](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.linregress.html)


- The linregress() function is part of the scipy.stats module and is used to perform linear regression analysis. It calculates several statistical values based on the input data.

- slope: The slope of the regression line. It represents the rate of change of the dependent variable (y) with respect to the independent variable (x) in a linear regression model.

- intercept: The y-intercept of the regression line. It represents the value of the dependent variable (y) when the independent variable (x) is zero in a linear regression model.

- r_value: The correlation coefficient (Pearson's r) between the independent and dependent variables. It measures the strength and direction of the linear relationship between the two variables. Values range from -1 to 1, where -1 indicates a perfect negative correlation, 1 indicates a perfect positive correlation, and 0 indicates no linear correlation.

- p_value: The p-value associated with the hypothesis test for the correlation coefficient. It provides an indication of the statistical significance of the observed correlation. Smaller p-values generally indicate stronger evidence against the null hypothesis of no correlation.

- std_err: The standard error of the estimate. It represents the standard deviation of the residuals (the differences between the observed and predicted values) in the linear regression model.
